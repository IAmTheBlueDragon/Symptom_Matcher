```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("./symptoms_machine.csv")
```

```python
figure =sns.barplot(x=df.columns[1:],y=df.sum(axis=0)[1:])
plt.xticks(range(len(df.columns[1:])), df.columns[1:], rotation='vertical')
figure.figure.set_size_inches(20,10)
```

```python
figure =sns.barplot(x=df['Diseases'].unique(),y=df['Diseases'].value_counts())
plt.xticks(range(len(df['Diseases'].unique())), df['Diseases'].unique(), rotation='vertical')
figure.figure.set_size_inches(20,10)
```

```python
X = df.drop('Diseases',axis=1)
y = df['Diseases']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.metrics import f1_score,classification_report,accuracy_score
```

```python
test_error = []

for n in range(1,200):
    # Use n random trees
    model = RandomForestClassifier(n_estimators=n,max_features='auto')
    model.fit(X_train,y_train)
    test_preds = model.predict(X_test)
    test_error.append(1-accuracy_score(test_preds,y_test))
```

```python
plt.plot(range(1,200),test_error,label='Test Error')
plt.legend()
```

```python
model = RandomForestClassifier(n_estimators=64,max_features='auto')
```

```python
model.fit(X_train,y_train)
```

```python
preds = model.predict(X_test)
```

```python
f1_score(y_test, preds, average='macro')
```

```python
print(classification_report(y_test,preds))
```
